```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as datetime
%matplotlib inline
color = sns.color_palette()
```

```python
df.columns = ['user_id', 'transaction_id', 'transaction_time', 'item_code',
       'item_description', 'quantity_purchased', 'unit_price', 'country']
```

# Data Cleaning

```python
df.info()
```

```python
#checking missing values for each column
df.isnull().sum().sort_values(ascending =False)
```

```python
#checking the rows with missing values
df[df.isnull().any(axis =1)].head()
```

```python
df['transaction_time'] = df.transaction_time.str.replace('IST ','')
df.head()
```

```python
df['transaction_time']  = pd.to_datetime(df['transaction_time'])
```

```python
#change item_description from upper case to lower case
df['item_description'] = df.item_description.str.lower()
```

```python
df.head()
```

```python
#removing rows with missing values
df_new = df.dropna()
df_new.isnull().sum().sort_values(ascending = False)
```

```python
df_new.info()
```

```python
df_new.describe().round(2)
```

```python
#removing rows with negative item_codes and user_ids 
df_new = df_new[df_new.user_id != -1]
df_new = df_new[df_new.item_code != -1]
df_new.describe().round(2)
```

```python
#removing rows with negative quantity_purchased
df_new = df_new[df_new.quantity_purchased > 0 ]
df_new.describe().round(2)
```

```python
#adding a new column - amount_spent
df_new['amount_spent'] = df_new['quantity_purchased'] * df_new['unit_price']
df_new.head()
```

```python
#rearranging the columns for ease 
df_new = df_new[['transaction_id','transaction_time','item_code','item_description','quantity_purchased','unit_price','amount_spent','user_id','country']]
df_new.head()
```

```python
df_new.insert(loc=2, column='year_month', value=df_new['transaction_time'].map(lambda x: 100*x.year + x.month))
df_new.insert(loc=3, column='month', value=df_new.transaction_time.dt.month)
df_new.insert(loc=4, column='day', value=(df_new.transaction_time.dt.dayofweek)+1)
df_new.insert(loc=5, column='hour', value=df_new.transaction_time.dt.hour)
df_new
```

```python
#removing the rows with transaction time past 2019-02-20
df_new = df_new[df_new.year_month < 201903]
df_new.transaction_time.max()
```

```python
#removing the outlier unit price since it dominates the further analysis
df_new = df_new[df_new.unit_price != df_new.unit_price.max()]
df_new.unit_price.max()
```

# RFM analysis

```python
#Restricting the data only to UNITED KINGDOM customers, which contains most of our customers historical data as the customer clusters may vary by geography
```

```python
df_uk = df_new[df_new.country == 'United Kingdom']
df_uk.head()
```

```python
df_uk=df_uk[['user_id','transaction_time','transaction_id','quantity_purchased','unit_price','amount_spent']]
df_uk.head()
```

```python
df_uk['transaction_time'].min(),df_uk['transaction_time'].max()
```

```python
now = datetime.date(2019,2,21)
now
```

```python
#RFM (Recency, Frequency, Monetary) analysis is a customer segmentation technique that uses past purchase behavior to divide customers into groups. 
# RECENCY (R): Days since last purchase
# FREQUENCY (F): Total number of purchases
# MONETARY VALUE (M): Total money this customer spent
```

```python
#recency
#To calculate recency, we need to choose a date point from which we evaluate how many days ago was the customer's last purchase.

df_uk['date'] = pd.DatetimeIndex(df_uk['transaction_time']).date
df_uk.head()
```

```python
recency_df = df_uk.groupby(by='user_id', as_index=False)['date'].max()
recency_df.columns = ['user_id','LastPurshaceDate']
recency_df.head()
```

```python
recency_df['recency'] = recency_df['LastPurshaceDate'].apply(lambda x: (now - x).days)
recency_df.drop('LastPurshaceDate',axis=1,inplace=True)
recency_df.head()
```

```python
#frequency
#Frequency helps us to know how many times a customer purchased from us
#To do that we need to check how many invoices are registered by the same customer.


df_ukcopy = df_uk
df_ukcopy.drop_duplicates(subset=['transaction_id', 'user_id'], keep="first", inplace=True)
#calculate frequency of purchases
frequency_df = df_ukcopy.groupby(by=['user_id'], as_index=False)['transaction_id'].count()
frequency_df.columns = ['user_id','frequency']
frequency_df.head()
```

```python
#monetary
#Monetary attribute answers the question: How much money did the customer spent over time?

monetary_df = df_uk.groupby(by='user_id',as_index=False).agg({'amount_spent': 'sum'})
monetary_df.columns = ['user_id','monetary']
monetary_df.head()
```

```python
temp_df = recency_df.merge(frequency_df,on='user_id')
rfm_df = temp_df.merge(monetary_df,on='user_id')
rfm_df.set_index('user_id',inplace=True)
rfm_df.head()
```

```python
quantiles = rfm_df.quantile(q=[0.25,0.5,0.75])
quantiles.to_dict()
```

```python
# Arguments (x = value, p = recency, monetary_value, frequency, d = quartiles dict)
def RScore(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1
# Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict)
def FMScore(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
```

```python
rfm= rfm_df
rfm['r_quartile'] = rfm['recency'].apply(RScore, args=('recency',quantiles,))
rfm['f_quartile'] = rfm['frequency'].apply(FMScore, args=('frequency',quantiles,))
rfm['m_quartile'] = rfm['monetary'].apply(FMScore, args=('monetary',quantiles,))
```

```python
rfm['RFMScore'] = rfm.r_quartile.map(str) \
                            + rfm.f_quartile.map(str) \
                            + rfm.m_quartile.map(str)
rfm.head()
```

```python
print("Best Customers: ",len(rfm[rfm['RFMScore']=='444']))
print('Loyal Customers: ',len(rfm[rfm['f_quartile']==4]))
print("Big Spenders: ",len(rfm[rfm['m_quartile']==4]))
print('Almost Lost: ', len(rfm[rfm['RFMScore']=='244']))
print('Lost Customers: ',len(rfm[rfm['RFMScore']=='144']))
print('Lost Cheap Customers: ',len(rfm[rfm['RFMScore']=='111']))
```
